Mini Project 1 : Social Media Survey
I wanted to use this survey to analyze the usage of social media by my baseball teammates and how it effected them mentally and possibly on the field. I used a small sample of fifteen guys. 

Q1: who? We will find the percentages of: Age, College Year, and gender. 

```python

import pandas as pd
import matplotlib.pyplot as plt

data = {'What is your age?': ['19', '23', '20', '19', '19', '19', '19', '20', '22', '21','22','21','20','20','19']}
df = pd.DataFrame(data)

age_percentage = df['What is your age?'].value_counts(normalize=True) * 100

plt.figure(figsize=(4, 4))
plt.pie(age_percentage, labels=age_percentage.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6'])
plt.title('Distribution of ages')

plt.show()
```

```python
data = {'What year of school are you in?': ['Freshman','Senior +','Junior','Freshman','Freshman','Freshman','Sophmore','Sophmore','Senior','Junior','Senior','Junior','Freshman','Sophmore','Sophmore']}
df = pd.DataFrame(data)

class_year_percentage = df['What year of school are you in?'].value_counts(normalize=True) * 100

colors = ['#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']

plt.figure(figsize=(4, 4))
plt.pie(class_year_percentage, labels=class_year_percentage.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Distribution of Class Years')

plt.show()

```

```python
data = {'What is your gender?': ['Male','Male','Male','Male','Male','Male','Male','Male','Male','Male','Male','Male','Male','Male','Male']}
df = pd.DataFrame(data)

gender_percentage = df['What is your gender?'].value_counts(normalize=True) * 100

colors = ['#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']

plt.figure(figsize=(4, 4))
plt.pie(gender_percentage, labels=gender_percentage.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Distribution of Gender')

plt.show()

```

A1: We can tell from these pie charts that most of our sample are 19, male and freshman in college.
